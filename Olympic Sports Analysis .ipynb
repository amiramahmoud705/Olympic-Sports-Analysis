#### Calculate the medals per each country for the best male and females athletes in all of the Summer editions and visualize the results

```python
df=summer.groupby(['Country','Medal'])['Gender'].count().reset_index().sort_values(by='Gender',ascending=False)
top_medals= df.pivot_table(index='Country', columns='Medal',fill_value=0)
top_medals
```

```python
top_10_countries=top_medals.sum(axis=1).nlargest(10)
top_10_countries=top_medals.loc[top_10_countries.index]
top_10_countries
```

```python
top_10_countries.plot.barh(figsize=(12,12),width=0.7,color=["darkgoldenrod", "gold", "LightGray"])
plt.title("Medals Distribution Of Top 10 Countries (Summer Olympics)")
plt.legend(title="Medal Type")
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 13))

# Ù…Ø¹Ø§Ù„Ø¬Ø© Ø¨ÙŠØ§Ù†Ø§Øª Ø§Ù„Ø±Ø¬Ø§Ù„
men = summer[summer['Gender'] == 'Men']
men = men.groupby(['Athlete', 'Medal'])['Country'].count().reset_index().sort_values(by='Country',ascending=False)
men = men.pivot_table(index='Athlete', columns='Medal', values='Country', fill_value=0)
top_men = men.loc[men.sum(axis=1).nlargest(15).index]
top_men.plot.barh(width=0.8, color=['darkgoldenrod', 'gold', 'LightGray'], ax=ax[0], legend=False)

ax[0].set_title('Best Male Athletes')
ax[0].set_ylabel('Athletes')
ax[0].set_xlabel('Number of medals')

women = summer[summer['Gender'] == 'Women']
women = women.groupby(['Athlete', 'Medal'])['Country'].count().reset_index().sort_values(by='Country',ascending=False)
women = women.pivot_table(index='Athlete', columns='Medal', values='Country', fill_value=0)
top_women = women.loc[women.sum(axis=1).nlargest(15).index] 
top_women.plot.barh(width=0.8, color=['darkgoldenrod', 'gold', 'LightGray'], ax=ax[1], legend=False)

ax[1].set_title('Best Female Athletes')
ax[1].set_ylabel('')
ax[1].set_xlabel('Number of medals')

plt.tight_layout()
plt.show()
```

```python
# Get all of the indicies of the old countries
old_countries = olympics[olympics.Country.isnull()].loc[1::].Code.value_counts().index
old_countries
```

```python
mapper = pd.Series(index=old_countries, name = "Country", data = ["Soviet Union", "East Germany", "Romania", "West Germany", "Czechoslovakia",
                               "Yugoslavia", "Unified Team", "Unified Team of Germany", "Mixed teams", "Serbia",
                              "Australasia", "Russian Empire", "Montenegro", "Trinidad and Tobago", "Bohemia", 
                              "West Indies Federation", "Singapore", "Independent Olympic Participants"])

mapper
```

```python
missing_indices = olympics[olympics.Country.isnull()].loc[1::].index
missing_indices
```

```python
olympics.loc[missing_indices, "Code"].map(mapper)
```

```python
olympics.Country.fillna(olympics.Code.map(mapper), inplace = True)
olympics.loc[missing_indices]
```

#### 3] Data Analysis & Visualization (EDA)

### What are the Top 10 Countries by total medals?

```python
Top_10_Countries_by_Medals= olympics.Country.value_counts().nlargest(10)
Top_10_Countries_by_Medals
```

```python
Top_10_Countries_by_Medals.plot(kind = "bar", fontsize = 16, figsize=(15,12) , color=('cyan'))
plt.title("Top 10 Countries by Medals", fontsize= 18)
plt.xlabel("Country",  fontsize= 18)
plt.ylabel("Medals" ,  fontsize= 18)
plt.show()
```

#### Split the total medals of Top 10 Countries into Summer / Winter. Are there typical Summer/Winter Games Countries?

```python
plt.figure(figsize=(20,10))
sns.set(font_scale=1.3, palette= "dark")
sns.countplot(data = olympics, x = "Country", color = ('darkviolet'), order = Top_10_Countries_by_Medals.index)
plt.xlabel("Country")
plt.ylabel("Total Medals")
plt.title("Top 10 Countries by Medals", fontsize = 20)
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.set(font_scale=1.5, palette= "dark")
sns.countplot(data = olympics, x = "Country", hue = "Edition", order = Top_10_Countries_by_Medals.index)
plt.title("Top 10 Countries by Medals", fontsize = 20)
plt.show()
```

#### Split the total medals of Top 10 Countries into Gold, Silver, Bronze

```python
plt.figure(figsize=(20,10))
sns.set(font_scale=1.5, palette= "dark")
sns.countplot(data = olympics, x = "Country", hue = "Medal", order =Top_10_Countries_by_Medals.index,
              hue_order = ["Gold", "Silver", "Bronze"], palette = ["gold", "Gray", "darkgoldenrod"])
plt.title("Top 10 Countries by Medals", fontsize = 20)
plt.show()
```
